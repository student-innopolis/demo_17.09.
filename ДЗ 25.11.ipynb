```python
import requests
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from collections import Counter
import nltk
```

```python
# Загрузка стоп-слов
nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))
```

```python
# URL для парсинга
url = "https://ria.ru/20241119/nalogovyy_vychet-1983843714.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Извлечение текста 
text = soup.find('body').get_text()
```

```python
text
```

```python
# Токенизация и очистка текста от стоп-слов
words = text.split()
cleaned_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
```

```python
cleaned_words
```

```python
# Подсчет частоты слов
word_counts = Counter(cleaned_words)
```

```python
word_counts
```

```python
# Находим 5 самых часто встречаемых слов
most_common_words = word_counts.most_common(5)
```

```python
most_common_words
```

```python
# Вывод результатов
print("5 самых часто встречаемых слов:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```
